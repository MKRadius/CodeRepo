## Problem 2

Using the *lineary spaced numerical array* of $x$ from $0$ to $10$ having $100$ samples, draw a graph of the function $y = (x-2)\cdot(x-8)$.

Annotate the graph:
- Add grid lines to the graph
- Add label for the x axis: "x-axis"
- add label for the y axis: "y-axis"
- Add title for the graph: "Polynomial function"



```python
x = np.linspace(0, 10, 100)
y = (x-2)*(x-8)

# Graphics
plt.plot(x, y)
plt.grid()

plt.title('Polynomial function')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

## Problem 3

Using linearly spaced numerical array of $x$ from $0$ to $10$ having $100$ samples draw in the same graph two polynomials $y1 = (x-2)\cdot(x-8)$ and $y2 = (x-2)\cdot(x-5)\cdot(x-8)$.

```python
x = np.linspace(0, 10, 100)
y1 = (x-2)*(x-8)
y2 = (x-2)*(x-5)*(x-8)

fig, ax = plt.subplots()

ax.plot(x, y1)
ax.plot(x, y2)
ax.grid(True)
```

## Problem 4

Using linearly spaced numerical array of $x$ from $0$ to $10$ having $100$ samples draw in the same graph two polynomials $y1 = (x-2)\cdot(x-8)$ and $y2 = (x-2)\cdot(x-5)\cdot(x-8)$. (same as in previous problem).

Annotate the graph:
- Add labels 'y1' and 'y2' to the plots.
- Add grid lines to the graph.
- Add x-label: 'x-axis'.
- Add y-label: 'y-axis'.
- Add title: 'Polynomial functions'.
- Add legend to the the graph.

```python
fig, ax = plt.subplots()

x = np.linspace(0, 10, 100)
y1 = (x-2)*(x-8)
y2 = (x-2)*(x-5)*(x-8)

ax.plot(x, y1, color = 'b', label = 'y1')
ax.plot(x, y2, color = 'r', label = 'y2')
ax.grid(True)
ax.set_title('Polynomial functions')
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.legend()
```

## Problem 5

Using linearly spaced numerical array of $x$ from $0$ to $10$ having $100$ samples calculate the values for the two polynomials $y1 = (x-2)\cdot(x-8)$ and $y2 = (x-2)\cdot(x-5)\cdot(x-8)$. (same as in previous problem).

Using two subplots, plot the first polynomial on the first axes and then the second polynomial on the second axes. 
- Use figure size of (6, 8).

Annotate the graph:
- Add grid lines to both plots.
- Add title: 'y1' to the first subplot.
- Add title: 'y2' to the second subplot.
- Add x-label: 'x-axis' to the second plot.

```python
fig, ax = plt.subplots(2, figsize = (6, 8))

x = np.linspace(0, 10, 100)
y1 = (x-2)*(x-8)
y2 = (x-2)*(x-5)*(x-8)

ax[0].plot(x, y1, color = 'b', label = 'y1')
ax[0].grid(True)
ax[0].set_title('y1')
ax[0].legend()

ax[1].plot(x, y2, color = 'r', label = 'y2')
ax[1].grid(True)
ax[1].set_title('y2')
ax[1].set_xlabel('x-axis')
ax[1].legend()
```

## Problem 6

- Calculate values for functions $f(x) = \sqrt{2x}$ and $g(x) = 5 - x$. Use linearly spaced numerical array of $x$ from $0$ to $10$ having 100 samples.
- Use two subplots and figure size of (6, 8).
- Draw the full graphs of both functions into the first axes. This part of the graph shows the full graphs.
- Draw the full graphs of both functions also into the second axes. This part is used for zooming in to the details.
- Add grid lines to both subplots.
- Change the x-axis and y-axis limits for the second axes, so that you can find graphically the intersection point (x0, y0) for the two equations (e.g. when $f(x) = g(x)$). Try to get the value minimum three significant figures accuracy.
- Read the intersection point (x0, y0) from the graph and draw it to the second graph using red circle.

```python
x = np.linspace(0, 10, 100)
f = np.sqrt(2*x)
g = 5 - x

fig, ax = plt.subplots(2, figsize = (6, 8))

ax[0].plot(x, f, color = 'b')
ax[0].plot(x, g, color = 'orange')
ax[0].grid(True)

ax[1].plot(x, f, color = 'b')
ax[1].plot(x, g, color = 'orange')
ax[1].grid(True)
ax[1].set_xlim(2.683, 2.684)
ax[1].set_ylim(2.316, 2.317)

# Intercation point
x0 = 2.683445
y0 = 2.316555
ax[1].plot(x0, y0, 'ro')
```
